```python
s = "asdfghjkl" 
t = "asdfghjk"
```

```python
def anagram(s:str,t:str):
    sorted_string_s = ''.join(sorted(s))
    print(sorted_string_s)
    sorted_string_t=''.join(sorted(t))
    print(sorted_string_t)
    i=0     #index for s string
    j=0     #index for t string
    p=0     #common index checking if all the characters of 2 strings are equal
    if len(sorted_string_s)==len(sorted_string_t):
        while i<len(sorted_string_s) and j<len(sorted_string_t):
            if sorted_string_s[i]==sorted_string_t[j]: 
                p=p+1
            else:
                break
            i=i+1
            j=j+1
    if p==len(sorted_string_s) and p==len(sorted_string_t):
        return True
    else:
        return False
```

```python
print(anagram(s,t))
    
```
